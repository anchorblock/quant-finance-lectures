```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from utils import *
```

## Correlation between stocks.

```python
datadir = "/root/codespace/quant-finance-lectures/amarstock_adjusted_data/daily"

print(get_top20_stocks(path=datadir))
```

Correlated stocks.

```python
stk1 = '00DS30'
stk2 = '00DSES'

df1 = get_prices(stk1, path=datadir) 
df2 = get_prices(stk2, path=datadir) 
returns1 = get_returns(dataframe=df1, field='close')
returns2 = get_returns(dataframe=df2, field='close')

plt.scatter(df1.close, df2.close)
plt.xlabel(stk1)
plt.ylabel(stk2)
plt.title('Stock prices from 2012 to 2022')
print("Correlation coefficient:", np.corrcoef(df1.close, df2.close)[0,1])
```

```python
rolling_correlation = df1.close.rolling(60).corr(df2.close)
ax = rolling_correlation.plot()
ax.set_xlabel('Day')
ax.set_ylabel('60-day Rolling Correlation')
```
