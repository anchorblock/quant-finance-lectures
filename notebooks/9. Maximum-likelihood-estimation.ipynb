```python
import math
import matplotlib.pyplot as plt
import numpy as np
import scipy
import scipy.stats
from utils import *
```

```python
datadir = "/root/codespace/quant-finance-lectures/amarstock_adjusted_data/daily"
print(get_top20_stocks(path=datadir))
```

```python
stockname1 = '00DS30'
df1 = get_prices(stockname1, path=datadir) 
# returns1 = get_returns(dataframe=df1, field='close')
mle_returns = np.diff(df1.close)/df1.close[:-1]

#now lets see this in a plot
mu, std = scipy.stats.norm.fit(mle_returns)
pdf = scipy.stats.norm.pdf
x = np.linspace(-1,1, num=100)
h = plt.hist(mle_returns, bins=x, density='true')
l = plt.plot(x, pdf(x, loc=mu, scale=std))
```
